## What's With all the zero values ?? 

```python
# Let's Investigate the frequency of zero for each feature and the response

print('Cement has value of zero at shape:',C[C['Cement'] == 0].shape)
print('Slag has value of zero at shape:',C[C['Slag'] == 0].shape)
print('FlyAsh has value of zero at shape:',C[C['FlyAsh'] == 0].shape)
print('Water has value of zero at shape:',C[C['Water'] == 0].shape)
print('Superplast has value of zero at shape:',C[C['Superplast'] == 0].shape)
print('Coarse_agg has value of zero at shape:',C[C['Coarse_agg'] == 0].shape)
print('Fine_agg has value of zero at shape:',C[C['Fine_agg'] == 0].shape)
print('Age has value of zero at shape:',C[C['Age'] == 0].shape)
      
print('CCS has value of zero at shape:',C[C['CCS'] == 0].shape)# Investigate the frequency of zero for each feature and the response
```

```python
# Time to Add Colums/Features to display whether Slag, FlyAsh, or superplasticizer are present in the concrete batch: 

C['Contains_Slag'] = C.Slag.apply(lambda Slag: Slag > 0)
C['Contains_FlyAsh'] = C.FlyAsh.apply(lambda FlyAsh: FlyAsh > 0)
C['Contains_Superplast'] = C.Superplast.apply(lambda Superplast: Superplast > 0)


# Encode categorical variables as integers.
C['Contains_Slag'] = pd.factorize(C.Contains_Slag)[0]
C['Contains_FlyAsh'] = pd.factorize(C.Contains_FlyAsh)[0]
C['Contains_Superplast'] = C.Contains_Superplast.astype('int64')


C.head()
```

```python
C.dtypes
```

```python
# Using boxplots, Let's see how the presence of Slag, FlyAsh and Superplast affect CCS. 

sns.factorplot(
    x='Contains_Slag',
    y='CCS',
    data=C,
    kind='box'
)

sns.factorplot(
    x='Contains_FlyAsh',
    y='CCS',
    data=C,
    kind='box'
)

sns.factorplot(
    x='Contains_Superplast',
    y='CCS',
    data=C,
    kind='box'
)
```

## Since the 'Age' feature in this dataset takes into account only 365 days of one unspecified year, we next must analyze this column and determine whether it could be useful or better off discarded. 


```python
#The majority of the value_counts for Age seem to be at the 28 day mark.  Let's visualize this with a simple bar chart:

C.Age.value_counts().plot(kind='bar') 
plt.xlabel('Age')
plt.ylabel('CCS');
```

```python
# The significance of the 28th day in the "Age" of concrete is that this marks the day that concrete is fully cured
C.Age.median()
```

```python
C.groupby('Age').CCS.mean().plot(kind='bar') 
plt.xlabel('Age')
plt.ylabel('CCS mean');
```

```python
# Let's see what the average Concrete Compressive Strenth value is along the Age timeline:
C.groupby('Age').CCS.mean().plot()
plt.ylabel('CCS', color = 'black');
```

```python
# Let's visualize this further in a boxplot where we can see outliers and where the means fall for each value of Age:

C.boxplot('CCS','Age',rot = 30,figsize=(10,10));
```

## We can see above that the Concrete Compressive Strength is much less during the first couple weeks, while the concrete is still curing. Let's split Age into two categories, at the median 28, representing the curing state of the concrete.  

# Let's create a new CuredConcrete column that indicates whether or not a row is cured_concrete, coded as 1, 
# or FreshConcrete, coded as 0--based on whether the Age is less than OR greater than/equal to the median: 28 days.  

# FreshConcrete: Age/days 1, 3, 7, and 14
# CuredConcrete: Age/days 28, 56, 90, 91, 100, 120, 180, 270, 360, and 365

```python
C['CuredConcrete'] = C.Age.map({1:0, 3:0, 7:0, 14:0, 28:1, 56:1, 90:1, 91:1, 100:1, 120:1, 180:1, 270:1, 360:1, 365:1})
```

```python
# Using boxplots, Let's see how the presence of Slag, FlyAsh and Superplast affect CCS. 

sns.factorplot(
    x='CuredConcrete',
    y='CCS',
    data=C,
    kind='box'
)
```
